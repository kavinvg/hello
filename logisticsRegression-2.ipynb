```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Creating sample Z points
z = np.arange(-5, 5, 0.1)

# Invoking Sigmoid function on all Z points
phi_z = sigmoid(z)

# Plotting the Sigmoid function
plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.xlabel('z')
plt.ylabel('$\phi(z)$')
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()

#Parameters:
#z: An array of values ranging from -5 to 5, used as input to the sigmoid function.
#phi_z: The output of the sigmoid function, which maps values of z to the range (0, 1).
#Applications:
#Classification: Logistic regression is widely used for binary classification tasks, where the goal is to assign inputs to one of two possible classes.
#Sigmoid Activation: Used to map any real-valued number into a probability value between 0 and 1.
#Decision Making: The sigmoid output helps decide the class by setting a threshold (e.g., 0.5).
#Theory:
#Logistic regression is a statistical method used for binary classification. It predicts the probability that a given input belongs to one of two classes (e.g., yes/no, true/false, 0/1). Here's a simple breakdown:

#1. Purpose:To classify data into one of two categories based on input features.
#Example: Classifying if an email is spam or not, based on certain features (words, length, etc.).
#2. How it works:
#Logistic regression uses a mathematical formula to predict the probability that an input belongs to a particular class.
#it uses a sigmoid function (or logistic function), which transforms any real-valued number into a value between 0 and 1.

#4. Decision Boundary:
#Logistic regression draws a decision boundary based on the features, separating the data into two classes.

#6. Why Not Linear Regression?
#Linear regression predicts continuous values, but logistic regression is specifically designed for binary outcomes.
#It restricts predictions to between 0 and 1, making it ideal for classification.
#7. Training the Model:
#Logistic regression uses maximum likelihood estimation (MLE) to find the best-fit parameters.
#The goal is to minimize the error between predicted probabilities and the actual labels.
#8. Applications:
#Medical diagnosis (e.g., predicting if a tumor is benign or malignant).
#Marketing (e.g., predicting if a customer will purchase a product).
#credit scoring (e.g., predicting if someone will default on a loan).
#In essence, logistic regression helps answer yes/no questions by estimating probabilities and making classifications based on those probabilities.






```
