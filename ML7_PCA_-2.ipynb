```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(0)
n_samples = 100
n_features = 5
```

```python
X = np.random.rand(n_samples, n_features)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k', s=50)
plt.title('PCA Result')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()


#1. Parameters Used
#n_samples = 100: You generated a random dataset with 100 samples (or data points).
#n_features = 5:each sample has 5 features or dimensions. This is the original dimensionality of the data.
#np.random.rand(n_samples, n_features):
#Generates a 100 x 5 matrix of random numbers between 0 and 1, which serves as your dataset.
#StandardScaler():
#This standardizes the data such that each feature has a mean of 0 and a standard deviation of 1. It's crucial for PCA since PCA is sensitive to the scale of the input features.
#PCA(n_components=2):
#This reduces the 5-dimensional data down to 2 principal components. The goal is to project the data onto a new feature space where the directions (components) capture the maximum variance in the data.
#plt.scatter():
#You plotted the PCA-transformed data. The plot shows how the samples look when represented by the first two principal components.

#2. Application of PCA
#PCA (Principal Component Analysis) has a variety of applications across different fields, including:
#Dimensionality Reduction: When working with high-dimensional data (like your 5-dimensional dataset), visualizing it or processing it can be computationally intensive or less informative. PCA reduces the dimensionality while preserving as much variance as possible. In this case, you reduced 5 dimensions to 2.
#Data Visualization: PCA helps in visualizing high-dimensional data in 2D or 3D for better understanding and pattern recognition. In this example, you've used PCA to visualize the data in 2D space using the first two principal components.
#Noise Reduction: PCA can be used to filter out noise by removing components with very low variance, which are assumed to be mostly noise.
#Feature Extraction: PCA can extract new features from a dataset that are uncorrelated and can improve performance in machine learning tasks.



#Principal Component Analysis (PCA) is a statistical technique used to simplify data by reducing its dimensions.
#Steps in Your Code Related to PCA Theory

#Data Standardization:
#Standardizing the features ensures that each feature contributes equally to the PCA (i.e., no feature dominates due to its larger scale).
#Covariance Matrix & Eigen Decomposition:
#The PCA algorithm calculates the covariance matrix of the standardized data and performs eigenvalue decomposition to obtain principal components.
#Projection onto Principal Components:
#The original 5-dimensional data is transformed and projected onto the first two principal components, which explain the most variance.
#Variance Explained by Each Component:
#PCA retains the components in order of their explained variance. The first component explains the most variance, followed by the second, and so on. In this example, the two components are likely capturing the majority of the dataâ€™s variability, allowing for effective dimensionality reduction.

```
